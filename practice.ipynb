```python
arr=[2,3,4,5,6,7,8]
```

```python
def largest(arr):
    maxele=arr[0]
    for i in arr:
        if i>maxele:
            maxele=i
    return maxele  
print(largest(arr))      
```

```python
def sumof(arr):
    total=0
    for i in arr:
        total+=i
    return total
print(sumof(arr))    
```

```python
#Frequency:
def frequency(arr):
    freq={}
    for i in arr:
        freq[i]=freq.get(i,0)+1
    return freq
print(frequency(arr))    
```

```python
def sorting(arr):
    for i in arr:
        if arr[i]>arr[i+1]:
            return False
    return True
print(sorting(arr))    
```

```python
def largest(arr):
    fir=0
    sec=0

    for i in arr:
        if i>fir:
            fir=i
    for i in arr:
        if i>sec and fir!=i:
            sec=i
    return sec
print(largest(arr))                
```

```python
def cop(arr):
    newa=[]
    for i in arr:
        newa.append(i)
    return newa
print(cop(arr))    
```

```python
arr=[1,2,3,4,5,6]
def linear(arr,k):
    n=len(arr)
    for i in range(0,n):
        if i==k:
            return i
    return -1
print(linear(arr,3))    
```

```python
def binary(arr,key):
    low=0
    high=len(arr)-1
    while low<=high:
        mid=((high+low)//2)
        if arr[mid]==key:
            return mid
        elif arr[mid]<key:
            low=mid+1
        else:    
            high=mid-1
    return
print(binary(arr,3))        
```

```python
arr=[1,2,2,3,3,4,5]
def rep(arr):
    seen=set()
    for i in arr:
        if i in seen:
            return i
        seen.add(i)
print(rep(arr))        
```

```python
arr=[1,1,2,3,4,2]
def dup(arr):
    n=len(arr)
    result=[]
    for i in range(n-1):
        if arr[i] in result:
            continue
        for j in range(i+1,n):
            if arr[i]==arr[j]:
                result.append(arr[i])
    return result
print(dup(arr))            
```
