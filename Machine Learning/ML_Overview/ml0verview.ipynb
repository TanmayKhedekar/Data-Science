```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#sample Datasets
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,0])

data = pd.DataFrame({"Feature": x.flatten(), "Target":y})
data
```

```python
#Data Visualization
plt.scatter(x,y)
plt.xlabel("Features")
plt.ylabel("Target")
plt.title("Simple linear relationship")
plt.show()
```

```python
#Train a sample ML Model
model=LinearRegression()
model.fit(x,y)
prediction=model.predict([6])
prediction
```

```python
### What just happened?
- We provided data to a machine learning model
- The model learned a pattern
- It predicted a new unseen value

This is the core idea of Machine Learning.
```

```python
## Key Takeaways
- ML learns patterns from data
- Models generalize to unseen data
- Data quality matters more than algorithms
```
