```python
# Unsupervised Learning
Clustering and dimensionality reduction using K-Means, DBSCAN, and PCA.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X, _ = make_blobs(
    n_samples=400,
    centers=4,
    cluster_std=1.2,
    random_state=42
)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels_kmeans = kmeans.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_kmeans)
plt.title("K-Means Clustering")
plt.show()
```

```python
inertia = []

for k in range(1, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(1, 10), inertia)
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_dbscan)
plt.title("DBSCAN Clustering")
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_kmeans)
plt.title("PCA + K-Means")
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
## Key Takeaways
- Unsupervised learning finds hidden patterns
- K-Means requires K and spherical clusters
- DBSCAN handles noise well
- PCA reduces dimensionality but loses interpretability
```
