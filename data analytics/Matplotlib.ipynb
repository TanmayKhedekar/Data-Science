```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 15, 30, 25])

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Trend")
plt.grid(True)
plt.show()
```

```python
#Line style
plt.plot(x,y,marker='o',linestyle="--")
plt.show()
```

```python
#Multiple lines on one plot:
y2 = [5, 15, 10, 20, 18]

plt.plot(x, y, label="Product A")
plt.plot(x, y2, label="Product B")
plt.legend()
plt.show()
```

```python
#Barchart:
products = ["A", "B", "C"]
sales = [100, 150, 80]

plt.bar(products, sales)
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product Sales")
plt.show()
```

```python
#Histogram:
data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title("Data Distribution")
plt.show()
```

```python
#Scatterplot:
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
