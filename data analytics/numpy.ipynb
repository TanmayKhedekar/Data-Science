```python
import numpy as np
a=np.array([1,2,3,4,5])
print(a)
```

```python
#2d array
a=np.array([[1,2],[3,4]])
print(a)
```

```python
#Array Properties
a.shape # size
a.size  #Total elements
a.dtype #Datatype
a.ndim #Dimensions
```

```python
np.eye(3) #Identity matrix
np.arange(1,10,2)
np.full((2,2),5)
np.ones((1,2))
np.zeros((2,2))
np.linspace(1,1,5)
```

```python
#Convert type:
arr = np.array([1, 2, 3], dtype=float)

arr.astype(int)
```

```python
#Indexing:
a=np.array([10,20,30,40,50,60])
a[1]
a[:4]
a[3:]
a[2:4]
a[-1]
```

```python
b = np.array([[1,2,3],[4,5,6]])

b[0,1]      # row 0, col 1
b[:,1]      # all rows, col 1
b[1,:]      # row 1
```

```python
#Boolean indexing:
a = np.array([10, 20, 30, 40])

a[a > 20]
a[(a > 10) & (a < 40)]
```

```python
#Math operations
a = np.array([1,2,3])
b = np.array([4,5,6])

a + b
a * b
a ** 2
```

```python
#Aggrigation functions:
a.sum()
a.mean()
a.min()
a.max()
a.std()
```

```python
#Reshaping
a=np.arange(1,7)
a.reshape(2,3)
a.flatten()
a.ravel()
```

```python
#Splitting
np.split(a, 3)
```

```python
#Set random:
np.random.rand(3,3)
np.random.randn(3,3)
np.random.randint(1, 100, size=(3,3))
```

```python
#searching and sorting
np.sort(a)
np.argsort(a)
np.where(a>2)
```

```python
#linear algebra:
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

np.dot(A, B)
np.linalg.det(A)
np.linalg.inv(A)
```

```python
#Handling missing values
a = np.array([1, 2, np.nan, 4])

np.isnan(a)
np.nanmean(a)
np.nan_to_num(a)
```

```python
# Python list
sum([i*i for i in range(1000000)])

# NumPy
arr = np.arange(1000000)
(arr ** 2).sum()
```
